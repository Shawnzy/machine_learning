# Load data

```python
# Import Data
import pandas as pd

CH = dict()
CH['title'] = 'Churn'
CH['RHC_data'] = pd.read_csv('./NN_OUTPUT/RHC_LOG.txt')
CH['SA_data'] = pd.read_csv('./NN_OUTPUT/SA0.55_LOG.txt')
CH['GA_data'] = pd.read_csv('./NN_OUTPUT/GA_100_50_50_LOG.txt')
CH['BP_data'] = pd.read_csv('./NN_OUTPUT/BACKPROP_LOG.txt')

problems = [CH]
```

# Learning curve - Plot Accuracy Vs. Iteration

```python
import matplotlib.pyplot as plt 

def learning_curve(prob):
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    ax = plt.subplot(111)
    plt.title(prob['title'] + ' Learning Curve (Accuracy)')
    plt.xlabel("Iterations")
    plt.ylabel("Accuracy")

    ax.plot('iteration', 'acc_tst', data=prob['RHC_data'].iloc[:100], label="RHC", color="#f33535")
    ax.plot('iteration', 'acc_tst', data=prob['SA_data'].iloc[:100], label="SA", color="#17b978")
    ax.plot('iteration', 'acc_tst', data=prob['GA_data'].iloc[:100], label="GA", color="#118df0")
    ax.plot('iteration', 'acc_tst', data=prob['BP_data'].iloc[:100], label="BP", color="#fd7014")

    legend = ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.3),
              ncol=4, fancybox=True, shadow=True)
    
 
    plt.savefig('./part1_plots/1_LC_Accuracy-{}.png'.format(prob['title']),bbox_inches="tight")
    plt.show()
    
    acc = dict()
    acc['RHA'] = round(prob['RHC_data'].acc_tst.iloc[-1],3)
    acc['SA'] = round(prob['SA_data'].acc_tst.iloc[-1],3)
    acc['GA']= round(prob['GA_data'].acc_tst.iloc[-1],3)
    acc['BP'] = round(prob['BP_data'].acc_tst.iloc[-1],3)
    print(acc)

for problem in problems:
    learning_curve(problem)
```

# Plot Mean Squared Error Vs. Iteration

# Plot Computation Time Vs. Iteration

```python
import matplotlib.pyplot as plt 

def learning_curve_time(prob):
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    ax = plt.subplot(111)
    plt.title(prob['title'] + ' Learning Curve (Computation Time)')
    plt.xlabel("Iterations")
    plt.ylabel("Computation Time (s)")

    ax.plot('iteration', 'elapsed', data=prob['RHC_data'].iloc[:100], label="RHC", color="#f33535")
    ax.plot('iteration', 'elapsed', data=prob['SA_data'].iloc[:100], label="SA", color="#17b978")
    #ax.plot('iteration', 'elapsed', data=prob['GA_data'].iloc[:100], label="GA", color="#118df0")
    ax.plot('iteration', 'elapsed', data=prob['BP_data'].iloc[:100], label="BP", color="#fd7014")

    legend = ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.3),
              ncol=4, fancybox=True, shadow=True)
    
 
    plt.savefig('./part1_plots/1_LC_CompTime-{}.png'.format(prob['title']),bbox_inches="tight")
    plt.show()
    time = dict()
    time['RHA'] = round(prob['RHC_data'].elapsed.iloc[-1],3)
    time['SA'] = round(prob['SA_data'].elapsed.iloc[-1],3)
    time['GA']= round(prob['GA_data'].elapsed.iloc[-1],3)
    time['BP'] = round(prob['BP_data'].elapsed.iloc[-1],3)
    print(time)

for problem in problems:
    learning_curve_time(problem)
```

## Fitness Vs Complexity

```python
def acc_comp(problem):
    complexities = []
    acc_vals = []
    acc_c = dict()
    for key in problem['files']:
        df = problem['files'][key]
        end_acc_val = round(df.acc_tst.iloc[-1],3)
        complexities.append(key)
        acc_vals.append(end_acc_val)
        acc_c[key] = end_acc_val
        
    fig = plt.figure()
    fig.set_size_inches(20, 5)
    plt.bar(complexities, acc_vals, align='center', alpha=.95, color='#118df0')
    #plt.xticks(y_pos, objects)
    plt.ylabel('Accuracy')
    plt.xlabel(problem['xlabel'])
    plt.title(problem['title']+' Accuracy Vs. Complexity Graph')
    plt.ylim(min(acc_vals)*.9, max(acc_vals)*1.1)
    plt.savefig('./part1_plots/1_CC_Accuracy_Hist-{}.png'.format(problem['saveas']),bbox_inches="tight")
    plt.show()
    print(acc_c)
for problem in problemsc:
    acc_comp(problem)
```

```python
def acc_comp_curve(problem):
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    ax = plt.subplot(111)
    plt.title(problem['title'] + ' Learning/Complexity Curve (Accuracy)')
    plt.xlabel("Iterations")
    plt.ylabel("Accuracy")
    for key in problem['files']:
        ax.plot('iteration', 'acc_trg', data=problem['files'][key].iloc[:300], label=key)

    legend = ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.3),
              ncol=5, fancybox=True, shadow=True)
    
 
    plt.savefig('./part1_plots/1_CC_Accuracy-{}.png'.format(problem['title']),bbox_inches="tight")
    plt.show()


for problem in problemsc:
    acc_comp_curve(problem)
```

```python
def time_comp(problem):
    complexities = []
    time_vals = []
    time_c = dict()
    for key in problem['files']:
        df = problem['files'][key]
        end_time_val = round(df.elapsed.iloc[-1],2)
        complexities.append(key)
        time_vals.append(end_time_val)
        time_c[key] = end_time_val
        
    fig = plt.figure()
    fig.set_size_inches(20, 5)
    plt.bar(complexities, time_vals, align='center', alpha=.95, color='#118df0')
    #plt.xticks(y_pos, objects)
    plt.ylabel('Computation Time (s)')
    plt.xlabel(problem['xlabel'])
    plt.title(problem['title']+' Computation Time Vs. Complexity Graph')
    plt.ylim(min(time_vals)*.9, max(time_vals)*1.1)
    plt.savefig('./part1_plots/1_CC_CompTime_Hist-{}.png'.format(problem['saveas']),bbox_inches="tight")
    plt.show()
    print(time_c)
for problem in problemsc:
    time_comp(problem)
```
