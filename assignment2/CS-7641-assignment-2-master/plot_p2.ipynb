# Load data

```python
# Import Data
import pandas as pd

FF = dict()
FF['title'] = 'FlipFlop'
FF['RHC_data'] = pd.read_csv('./FLIPFLOP/FLIPFLOP_RHC_1_LOG.txt')
FF['SA_data'] = pd.read_csv('./FLIPFLOP/FLIPFLOP_SA0.55_1_LOG.txt')
FF['GA_data'] = pd.read_csv('./FLIPFLOP/FLIPFLOP_GA100_30_50_3_LOG.txt')
FF['MIMIC_data'] = pd.read_csv('./FLIPFLOP/FLIPFLOP_MIMIC100_50_0.9_1_LOG.txt')

TS = dict()
TS['title'] = 'Traveling_Salesman'
TS['RHC_data'] = pd.read_csv('./TSP/TSP_RHC_1_LOG.txt')
TS['SA_data'] = pd.read_csv('./TSP/TSP_SA0.55_1_LOG.txt')
TS['GA_data'] = pd.read_csv('./TSP/TSP_GA100_50_30_2_LOG.txt')
TS['MIMIC_data'] = pd.read_csv('./TSP/TSP_MIMIC100_50_0.9_2_LOG.txt')

CP = dict()
CP['title'] = 'Continuous_Peaks'
CP['RHC_data'] = pd.read_csv('./CONTPEAKS/CONTPEAKS_RHC_1_LOG.txt')
CP['SA_data'] = pd.read_csv('./CONTPEAKS/CONTPEAKS_SA0.55_1_LOG.txt')
CP['GA_data'] = pd.read_csv('./CONTPEAKS/CONTPEAKS_GA100_30_50_1_LOG.txt')
CP['MIMIC_data'] = pd.read_csv('./CONTPEAKS/CONTPEAKS_MIMIC100_50_0.7_1_LOG.txt')

problems = [FF, TS, CP]
```

# Learning curve - Plot Fitness Vs. Iteration

```python
import matplotlib.pyplot as plt 

def learning_curve(prob):
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    ax = plt.subplot(111)
    plt.title(prob['title'] + ' Learning Curve')
    plt.xlabel("Iterations")
    plt.ylabel("Fitness Score")

    ax.plot('iterations', 'fitness', data=prob['RHC_data'], label="RHC", color="#f33535")
    ax.plot('iterations', 'fitness', data=prob['SA_data'], label="SA", color="#17b978")
    ax.plot('iterations', 'fitness', data=prob['GA_data'], label="GA", color="#118df0")
    ax.plot('iterations', 'fitness', data=prob['MIMIC_data'], label="MIMIC", color="#fd7014")

    legend = ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.3),
              ncol=4, fancybox=True, shadow=True)
    
 
    plt.savefig('./part2_plots/learning_curve-{}.png'.format(prob['title']),bbox_inches="tight")
    plt.show()

for problem in problems:
    learning_curve(problem)
```

# Plot Computation time Vs. Iteration

```python
def comp_curve(prob):
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    ax = plt.subplot(111)
    plt.title(prob['title'] + ' Computation Time Curve')
    plt.xlabel("Iterations")
    plt.ylabel("Computation Time (s)")

    ax.plot('iterations', 'time', data=prob['RHC_data'], label="RHC", color="#f33535")
    ax.plot('iterations', 'time', data=prob['SA_data'], label="SA", color="#17b978")
    ax.plot('iterations', 'time', data=prob['GA_data'], label="GA", color="#118df0")
    #ax.plot('iterations', 'time', data=prob['MIMIC_data'], label="MIMIC", color="#fd7014")

    legend = ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.3),
              ncol=4, fancybox=True, shadow=True)
    
 
    #fig.savefig(bbox_extra_artists=(legend,))
    plt.savefig('./part2_plots/comp_time_curve-{}.png'.format(prob['title']),bbox_inches="tight")
    plt.show()
    
    time = dict()
    time['RHA'] = round(prob['RHC_data'].time.iloc[-1],3)
    time['SA'] = round(prob['SA_data'].time.iloc[-1],3)
    time['GA']= round(prob['GA_data'].time.iloc[-1],3)
    time['MIMIC'] = round(prob['MIMIC_data'].time.iloc[-1],3)
    print(time)


for problem in problems:
    comp_curve(problem)
```

## Fitness Vs Complexity

```python
def fit_comp(problem):
    complexities = []
    fit_vals = []
    for key in problem['files']:
        df = problem['files'][key]
        end_fit_val = df.fitness.iloc[-1]
        complexities.append(key)
        fit_vals.append(end_fit_val)
        
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    plt.bar(complexities, fit_vals, align='center', alpha=.95, color='#118df0')
    #plt.xticks(y_pos, objects)
    plt.ylabel('Fitness Score')
    plt.xlabel(problem['xlabel'])
    plt.title(problem['title']+' Fitness Score Vs. Complexity Graph')
    plt.ylim(min(fit_vals)*.9, max(fit_vals)*1.1)
    plt.savefig('./part2_plots/Complexity_fit_curve-{}.png'.format(problem['saveas']),bbox_inches="tight")
    plt.show()

for problem in problemsc:
    fit_comp(problem)
```

## Flipflop Computation Time Vs Complexity

```python
def time_comp(problem):
    complexities = []
    time_vals = []
    for key in problem['files']:
        df = problem['files'][key]
        end_time_val = df.time.iloc[-1]
        complexities.append(key)
        time_vals.append(end_time_val)
        
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    plt.bar(complexities, time_vals, align='center', alpha=.95, color='#118df0')
    #plt.xticks(y_pos, objects)
    plt.ylabel('Computational Time (s)')
    plt.xlabel(problem['xlabel'])
    plt.title(problem['title']+' Computation Time Vs Complexity Graph')
    plt.ylim(min(time_vals)*.9, max(time_vals)*1.1)
    plt.savefig('./part2_plots/Complexity_time_curve-{}.png'.format(problem['saveas']),bbox_inches="tight")
    plt.show()

for problem in problemsc:
    time_comp(problem)
```
